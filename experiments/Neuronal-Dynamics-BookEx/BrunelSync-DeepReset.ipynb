```python

%matplotlib inline
from neurodynex3.brunel_model import LIF_spiking_network
from neurodynex3.tools import plot_tools
import brian2 as b2
```

```python

ft = 20.0
for rv in [
    0.0, 1.0, 2.0, 5.0, 7.0, 8.0, 9.0, 
    10.0, 11.0, 12.0, 13.0, 15.0, 18.0, 20.0, 
]:

    rate_monitor, spike_monitor, voltage_monitor, monitored_spike_idx = LIF_spiking_network.simulate_brunel_network(
        firing_threshold= ft * b2.mV,
        v_reset= rv * b2.mV,

        N_Excit=10_000,
        N_Inhib=2_500,
        N_extern=1_000,
        g=3., poisson_input_rate=50 *b2.Hz,
        sim_time=100. * b2.ms,
    )
    plot_tools.plot_network_activity(
        rate_monitor, spike_monitor, voltage_monitor, spike_train_idx_list=monitored_spike_idx,
        t_min=0.*b2.ms,
        sup_title= f'rv={rv}, ft={ft}',
    )
```
